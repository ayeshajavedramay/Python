```python
print("\n\n TASK 2 : \n")


class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height



results = {}

data_sets = [
    (10, 8, 5, 6, 4),
    (12, 6, 7, 5, 3),
    (15, 10, 8, 9, 6)
]

for i, (a, b, h, base, height) in enumerate(data_sets, start=1):
    trap = Trapezoid(a, b, h)
    para = Parallelogram(base, height)

    trap_area = trap.area()
    para_area = para.area()

    results[f"Set {i}"] = {
        "Trapezoid": trap_area,
        "Parallelogram": para_area
    }

print("Areas for different sets:")
for key, value in results.items():
    print(f"\n{key}:")
    for shape, area in value.items():
        print(f"{shape} Area = {area}")

max_shape = ""
max_area = 0

for dataset in results.values():
    for shape, area in dataset.items():
        if area > max_area:
            max_area = area
            max_shape = shape

print("\nShape with maximum area:", max_shape)
print("Maximum area:", max_area)
```
